```python
def double_list(my_list):
    return [x * 2 for x in my_list]

# Example
my_list = [1, 2, 3, 4, 5]
result_list = double_list(my_list)
print("List Output:", result_list)
```

```python

import numpy as np

def double_array(my_array):
    return my_array * 2
# Example
my_array = np.array([1, 2, 3, 4, 5])
result_array = double_array(my_array)
print("Array Output:", result_array)
```
